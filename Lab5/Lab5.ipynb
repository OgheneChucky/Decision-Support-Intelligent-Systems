```python
import pandas as pd
```

```python
HousingData = pd.read_csv('kc_house_data.csv')
HousingData
```

```python
import numpy as np
HousingData.select_dtypes(np.number)
```

```python
HousingData.isnull().sum()
```

```python
HousingData.drop(['id','date','waterfront','view','condition','grade','sqft_above','sqft_basement','yr_built','yr_renovated','zipcode','lat','long','sqft_living15','sqft_lot15'], axis=1, inplace=True)
```

```python
HousingData.head()
```

```python
X = HousingData.drop (['price'], axis = 1)
Y = HousingData['price']
```

```python
X_train = np.array(X[0:int(0.80*len(X))])
Y_train = np.array(Y[0:int(0.80*len(Y))])
X_test = np.array(X[int(0.80*len(X)):])
Y_test = np.array(Y[int(0.80*len(Y)):])
len(X_train), len(Y_train), len(X_test), len(Y_test)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
```

```python
LR_fit = LR.fit(X_train, Y_train)
```

```python
LR_pred = LR_fit.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Logistic Regression is %f percent accurate" % (accuracy_score(LR_pred, Y_test)*100))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
seed=0
cv = KFold(n_splits=5,random_state=2, shuffle=True)
```

```python
def return_score(model,X_train, X_test, y_train, y_test):
   model.fit(X_train, y_train)
   score = model.score(X_test, y_test)
   return score
scores = []
```

```python
model = LogisticRegression()
for train_index, test_index in cv.split(X,Y):
   X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
   score = return_score(model,X_train, X_test, y_train, y_test)   
   scores.append(score)
print("Accuracy score in each iteration: {}".format(scores))
print("K-Fold Score: {}".format(np.mean(scores)))
```
